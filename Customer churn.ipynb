33% is very high. But we don't really need to use all columns in our data. Because the is very weak correlation with some columns.

Lets get rid of them and use only columns that has got strong correlation.

```python
data = df[['Churn','Tenure','Complain','DaySinceLastOrder','CashbackAmount','MaritalStatus','Gender']]
```

```python
print(f"Number of nulls: {data.isnull().sum().sum()}")

missing_rows=data[data.isnull().any(axis=1)].shape[0]

print(f"Missing rows: {np.round(missing_rows/len(data)*100)}%")
```

```python
data=data.dropna()
data.shape
```

```python
data["Churn"].value_counts()/len(data)
```

## Preperation to ML

```python
data.info()
```

```python
# Categorical columns are converted to int
encoded = pd.get_dummies(data,dtype=int)
encoded.head()
```

```python
X=encoded.drop("Churn",axis=1)
y=encoded["Churn"]
```

```python
scaler=StandardScaler()
X=scaler.fit_transform(X)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=0,)
```

```python
LR_model=LogisticRegression()
LR_model.fit(X_train,y_train)
```

```python
#assess the model
y_pred=LR_model.predict(X_test)
print(metrics.classification_report(y_test,y_pred))
print("Model accuracy: ",metrics.accuracy_score(y_test,y_pred))

#confusion matrix
conf_mat=metrics.confusion_matrix(y_test,y_pred)
sns.heatmap(conf_mat,annot=True,fmt="g")
plt.show()
```

## Decision Tree

```python
tree_model=DecisionTreeClassifier()
tree_model.fit(X_train,y_train)
```

```python
y_pred=tree_model.predict(X_test)

print("Classification report:", metrics.classification_report(y_test,y_pred))
print("Model accuracy: ", metrics.accuracy_score(y_test,y_pred))

conf_mat=metrics.confusion_matrix(y_test,y_pred)
sns.heatmap(conf_mat,annot=True,fmt="g")
plt.show()
```

```python
cols=encoded.drop(columns="Churn").columns

plt.figure(figsize=(30,20))
plot_tree(tree_model,feature_names=cols,filled=True)
plt.show()
```

## XGBoost

```python

xgb_model = XGBClassifier()
xgb_model.fit(X_train, y_train)

y_pred = xgb_model.predict(X_test)
print(metrics.classification_report(y_test, y_pred))
print("Model accuracy:", metrics.accuracy_score(y_test,y_pred))


conf_mat = metrics.confusion_matrix(y_test, y_pred)
sns.heatmap(conf_mat, annot=True,fmt="g")
plt.show()
```
